Time Series Forecasting for Bike Demand

Why do we need time series forecasting?
- Current LP uses historical averages
- Real operations require future demand estimates

Objective:
- Forecast hourly bike demand to serve as input to the optimization model

Data Used:
- hour.csv (present in data folder)
- target variable cnt (total number of bikes rented)
- cnt = registered + casual
- Difference between registered and casual is registered is the number of registered users renting bikes; Interpreation: more consistant usage used to measure hourly demand; Whereas casual users are bikes rented that doesn't have an account; Interpretation: less consistant than registered users
